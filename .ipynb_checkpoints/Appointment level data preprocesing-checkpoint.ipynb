```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/Case Rigshospitalet.csv')
```

```python
from utils import column_name_mapping
# renaming columns to english
df = df.rename(columns=column_name_mapping)
```

```python
df.info()
df.head()
```

```python
# Basic data exploration
summary_stats = {
    "Number of Patients": df["patientID"].nunique(),
    "Number of Appointments": df["patientContactID"].nunique(),
    "Number of Total Meetings": df["treatmentContactID"].nunique(),
    "Number of Female Patients": df[df["patientGender"] == "Kvinde"]["patientID"].nunique(),
    "Number of Male Patients": df[df["patientGender"] == "Mand"]["patientID"].nunique(),
    "Number of Unique Diseases (Primary)": df["primaryDiagnosisCode"].nunique(),
    "Number of Unique Diseases (Secondary)": df["secondaryDiagnosisCode"].nunique(),
    "Number of Unique Countries": df["patientCountry"].nunique(),
    "Number of Unique Regions": df["patientRegion"].nunique(),
    "Number of Unique Municipalities": df["patientMunicipality"].nunique(),
    "Number of Unique Postal Codes": df["patientPostalCode"].nunique(),
}

# Convert summary statistics into a DataFrame for better visualization
summary_df = pd.DataFrame(list(summary_stats.items()), columns=["Metric", "Value"])

# Display the summary table
print(summary_df)
```

```python
# Analyze missing values
missing_values = df.isnull().sum()  # Count of missing values per column
missing_percentage = (missing_values / len(df)) * 100  # Percentage of missing values

# Create a DataFrame to display the missing values analysis
missing_df = pd.DataFrame({
    "Feature": df.columns,
    "Missing Values": missing_values,
    "Percentage Missing (%)": missing_percentage
}).sort_values(by="Percentage Missing (%)", ascending=False)  # Sort by highest missing percentage

# Display the missing values analysis
print(missing_df)
```

```python
columns_to_drop = ["secondaryDiagnosisCode", "procedureSupplementaryCodes"]
df_cleaned = df.drop(columns=columns_to_drop)
```

```python
# Step 1: Count unique `treatmentContactID` per `patientContactID`
meeting_counts = df.groupby("patientContactID")["treatmentContactID"].nunique().reset_index()
meeting_counts.rename(columns={"treatmentContactID": "Meetings Num"}, inplace=True)

# Step 2: Group by `patientContactID` (appointment level)
df_appointments = df.groupby("patientContactID").agg({
    "patientID": "first",  # Keep the first patient ID (all rows belong to the same patient)
    "patientAgeAtContactStart": "first",  # Age remains the same
    "patientGender": "first",  # Gender remains the same
    "appointmentUrgency": "first",  # Urgency remains the same
    "contactType": "first",  # Contact type remains the same
    "primaryDiagnosisCode": "first",  # Keep the first primary diagnosis
    "contactStartDate": "first",  # Keep the first occurrence (will convert to datetime later)
    "contactEndDate": "first",  # Keep the first occurrence (will convert to datetime later)
    "visitType": "first",  # Keep the first visit type
    "patientMaritalStatus": "first",  # Keep marital status
    "patientCountry": "first",  # Keep country of residence
    "patientRegion": "first",  # Keep region of residence
    "patientPostalCode": "first",  # Keep postal code
    "patientMunicipality": "first",  # Keep municipality
}).rename(columns={"treatmentContactID": "NumberOfMeetings"})  # Rename column for clarity

# Step 3: Merge the correct Meetings Num count
df_appointments = df_appointments.merge(meeting_counts, on="patientContactID", how="left")
```

```python
df_appointments.info()
df_appointments.head()
```

```python
# Step 2: Convert `contactStartDate` and `contactEndDate` to datetime (now with fewer rows!)
df_appointments["contactStartDate"] = pd.to_datetime(df_appointments["contactStartDate"])
df_appointments["contactEndDate"] = pd.to_datetime(df_appointments["contactEndDate"])

# Step 3: Calculate appointment duration in minutes
df_appointments["AppointmentDuration(min)"] = (df_appointments["contactEndDate"] - df_appointments["contactStartDate"]).dt.total_seconds() / 60

# Drop `contactStartDate` and `contactEndDate` if not needed anymore
df_appointments.drop(columns=["contactStartDate", "contactEndDate"], inplace=True)

# Print a preview of the transformed dataset
df_appointments.head()
```

```python
# Convert `patientGender` to binary: Female = 0, Male = 1
df_appointments["patientGender"] = df_appointments["patientGender"].map({"Kvinde": 0, "Mand": 1})

# Create dummy variables for `visitType`
df_appointments = pd.get_dummies(df_appointments, columns=["contactType"], prefix="contactType")

df_appointments.head()
```

```python
# Step 4: Translate Dummy Column Names to English
translation_dict = {
    "contactType_Fysisk fremmøde": "Physical Appointment",
    "contactType_Virtuel pt.kt.": "Virtual Appointment",
    "contactType_Udekontakt": "HomeVisit Appointment",
}

df_appointments.rename(columns=translation_dict, inplace=True)
```

```python
df_appointments.head()
```

```python
# Ensure dummy variables are binary (0/1) instead of True/False
dummy_columns = ["Physical Appointment", "Virtual Appointment", "HomeVisit Appointment"]
df_appointments[dummy_columns] = df_appointments[dummy_columns].astype(int)
```

```python
# Define the column name mapping
rename_dict = {
    "patientContactID": "AppointmentID",
    "patientAgeAtContactStart": "Age",
    "patientGender": "Gender",
    "primaryDiagnosisCode": "Diagnosis Code",
    "NumberOfMeetings": "Meetings Num",
    "visitType": "Type of visit",
    "patientMaritalStatus": "Marital Status",
    "patientCountry": "Country",
    "patientRegion": "Region",
    "patientPostalCode": "Postal Code",
    "patientMunicipality": "Municipality",
    "appointmentUrgency": "Urgency",
}

# Apply the renaming
df_appointments.rename(columns=rename_dict, inplace=True)
```

```python
# Convert 'Urgency' to binary: 0 = No Urgency (Planlagt), 1 = Urgency (Akut)
df_appointments["Urgency"] = df_appointments["Urgency"].map({"Akut": 1, "Planlagt": 0})
df_appointments.head()
```

```python
# save to csv
df_appointments.to_csv('data/Appointments_dataframe.csv', index=False)
```

```python
df_diagnosis = pd.read_csv('data/Translated diagnosis dimension table.csv')
df_diagnosis.info()
df_diagnosis.head()
```

```python
# Step 1: Merge `df_appointments` with `df_diagnosis`
df_appointments = df_appointments.merge(
    df_diagnosis[["Diagnosis", "Disease Name", "Group 1", "Group 2", "Group 3"]],  # Selecting relevant columns
    left_on="Diagnosis Code",  # Matching column in `df_appointments`
    right_on="Diagnosis",  # Matching column in `df_diagnosis`
    how="left"  # Left join to keep all appointments
)

# Step 2: Drop the redundant "Diagnosis" column from df_diagnosis
df_appointments.drop(columns=["Diagnosis"], inplace=True)

# Step 3: Reorder columns (placing Disease Name & Group 1 after Diagnosis Code)
desired_order = (
    df_appointments.columns[:df_appointments.columns.get_loc("Diagnosis Code") + 1].tolist() + 
    ["Disease Name", "Group 1", "Group 2", "Group 3"] + 
    [col for col in df_appointments.columns if col not in ["Diagnosis Code", "Disease Name", "Group 1", "Group 2", "Group 3"]]
)

df_appointments = df_appointments[desired_order]

# Step 4: Save the updated dataset (optional)
df_appointments.to_csv("data/Processed_Appointments_With_Diseases.csv", index=False)

# Step 5: Verify the new column order
df_appointments.head()
```
