```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.utils import resample
import seaborn as sns
import plotly.express as px
import plotly.io as pio
from matplotlib.colors import to_hex
from sklearn.mixture import GaussianMixture
```

```python
datapath = 'data/patient_dataframe.csv'
data = pd.read_csv(datapath)
```

```python
data.head()
```

```python
len(data)
```

```python
# normalize columns by number of visits (helps with correlation)
data['numDiagnosis'] = data['Diagnosis'].apply(len)
data['avgProcedures'] = data['numProcedures']/data['numVisits']
data['avgContacts'] = data['numContacts']/data['numVisits']
data['avgDuration'] = data['cumulativeVisitDuration']/data['numVisits']
data['PhysicalAppointmentsRatio'] = data['PhysicalAppointments']/data['numVisits']
data['HomeAppointmentsRatio'] = data['HomeAppointments']/data['numVisits']
data['VirtualAppointmentsRatio'] = data['VirtualAppointments']/data['numVisits']
data['AcuteRatio'] = data['numAcute']/data['numVisits']
data['PlannedRatio'] = data['numPlanned']/data['numVisits']
data['RegularOutpatientRatio'] = data['numRegularOutpatient']/data['numVisits']
data['AcuteAmbulatoryRatio'] = data['numAcuteAmbulatory']/data['numVisits']
data['UnderObservationRatio'] = data['numUnderObservation']/data['numVisits']
```

```python
def gap_statistic(k, data, B=10):
    # Perform KMeans on the actual data
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(data)
    W_k = kmeans.inertia_  # Inertia for the actual data
    
    # Generate reference (null) data (uniform distribution)
    ref_dispersion = []
    for _ in range(B):
        ref_data = np.random.uniform(low=np.min(data, axis=0), high=np.max(data, axis=0), size=data.shape)
        kmeans.fit(ref_data)
        ref_dispersion.append(kmeans.inertia_)
    
    # Compute Gap Statistic
    gap = np.log(np.mean(ref_dispersion)) - np.log(W_k)
    return gap
```

## K-means

```python
clustering_columns = [
'numVisits', 'Age', 'Gender', 'numDiagnosis', 'avgProcedures',
'avgContacts', 'avgDuration', 'PhysicalAppointmentsRatio',
'HomeAppointmentsRatio', 'VirtualAppointmentsRatio', 'AcuteRatio',
'PlannedRatio', 'RegularOutpatientRatio', 'AcuteAmbulatoryRatio',
'UnderObservationRatio'    
]

clustering_df = data[clustering_columns]
```

```python
# standardize
scaler = StandardScaler()
df_scaled = scaler.fit_transform(clustering_df)

# sample data to spped up silhouette
sample_data = resample(df_scaled, n_samples=10000, random_state=42)

# clustering
inertia = []
silhouette_scores = []
gap_statistics = []
K_range = range(2, 11)

for k in K_range:
    print('Running ', k)
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

    # calculate score on sample to speed it up
    score = silhouette_score(sample_data, kmeans.fit_predict(sample_data), metric='euclidean')
    silhouette_scores.append(score)

    # Calculate the Gap Statistic for the current K
    gap = gap_statistic(k, df_scaled)
    gap_statistics.append(gap)
```

```python
# Plot the results: Inertia, Silhouette Score, and Gap Statistic
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot Inertia (WCSS)
axes[0].plot(K_range, inertia, marker='o', linestyle='-')
axes[0].set_xlabel('Number of Clusters (K)')
axes[0].set_ylabel('Inertia (WCSS)')
axes[0].set_title('Elbow Method for Optimal K')
axes[0].grid()

# Plot Silhouette Score
axes[1].plot(K_range, silhouette_scores, marker='o', linestyle='-')
axes[1].set_xlabel('Number of Clusters (K)')
axes[1].set_ylabel('Silhouette Score')
axes[1].set_title('Silhouette Score for Optimal K')
axes[1].grid()

# Plot Gap Statistic
axes[2].plot(K_range, gap_statistics, marker='o', linestyle='-')
axes[2].set_xlabel('Number of Clusters (K)')
axes[2].set_ylabel('Gap Statistic')
axes[2].set_title('Gap Statistic for Optimal K')
axes[2].grid()

# Adjust layout
plt.tight_layout()
plt.show()
```

## Gaussian Mixture Model

```python
clustering_columns = [
'numVisits', 'Age', 'Gender', 'numDiagnosis', 'avgProcedures',
'avgContacts', 'avgDuration', 'PhysicalAppointmentsRatio',
'HomeAppointmentsRatio', 'VirtualAppointmentsRatio', 'AcuteRatio',
'PlannedRatio', 'RegularOutpatientRatio', 'AcuteAmbulatoryRatio',
'UnderObservationRatio'    
]

clustering_df = data[clustering_columns]
```

```python
# standardize
scaler = StandardScaler()
df_scaled = scaler.fit_transform(clustering_df)

# sample data to spped up silhouette
sample_data = resample(df_scaled, n_samples=10000, random_state=42)
```

```python
k_range = range(2, 11)
BIC = []
AIC = []
silhouette_scores = []

for k in k_range:
    print(f'Running k={k}')
    gmm = GaussianMixture(n_components=k, covariance_type='full', max_iter=1000, reg_covar=0.1, random_state=42)
    gmm.fit(df_scaled)
    BIC.append(gmm.bic(df_scaled))
    AIC.append(gmm.aic(df_scaled))
    labels = gmm.predict(sample_data)
    silhouette_scores.append(silhouette_score(sample_data, labels, metric='euclidean'))
```

```python
# Create subplots
fig, axes = plt.subplots(2, 1, figsize=(7, 7))

# AIC/BIC
axes[0].plot(k_range, BIC, marker='o', label='BIC')
axes[0].plot(k_range, AIC, marker='s', label='AIC')
axes[0].set_xlabel('Number of Clusters (k)')
axes[0].set_ylabel('BIC/AIC Score')
axes[0].legend()
axes[0].grid(True)

# Silhouette
axes[1].plot(k_range, silhouette_scores, marker='^', label='Silhouette', c='green')
axes[1].set_xlabel('Number of Clusters (k)')
axes[1].set_ylabel('Silhouette Score')
axes[1].legend()
axes[1].grid(True)

# Adjust layout
plt.suptitle('Scores evaluation for different number of clusters')
plt.tight_layout()
plt.show()
```

## Map patients to cluster labels

Decided to use 3 different clusterings. Hard to say which one is the best. We should determine it after checking the results since the evaluation metrics are not of much help:
- K-means / subset of features / k = 5
- GMM / subset of features / k = 5
- GMM / all features / k = 4 

By subset of features I mean features left out after I eliminated some of them due to weird distribution. See below.

```python
all_features = [
'numVisits', 'Age', 'Gender', 'numDiagnosis', 'avgProcedures',
'avgContacts', 'avgDuration', 'PhysicalAppointmentsRatio',
'HomeAppointmentsRatio', 'VirtualAppointmentsRatio', 'AcuteRatio',
'PlannedRatio', 'RegularOutpatientRatio', 'AcuteAmbulatoryRatio',
'UnderObservationRatio'    
]

clustering_full = data[all_features]

subset_of_features = [
'numVisits', 'Age', 'Gender', 'numDiagnosis', 'avgProcedures',
'avgContacts', 'avgDuration', 'PhysicalAppointmentsRatio',
 'VirtualAppointmentsRatio', 'PlannedRatio', 'UnderObservationRatio'    
]

clustering_subset = data[subset_of_features]
```

```python
# standardize
scaler = StandardScaler()
df_scaled_full = scaler.fit_transform(clustering_full)
df_scaled_subset = scaler.fit_transform(clustering_subset)
```

```python
# GMM with subset of features and 5 clusters
gmm_5 = GaussianMixture(n_components=5, covariance_type='full', max_iter=1000, reg_covar=0.1, random_state=42)
gmm_5.fit(df_scaled_subset)
labels_gmm_5 = gmm_5.predict(df_scaled_subset)

# GMM with full features and 4 clusters
gmm_4 = GaussianMixture(n_components=4, covariance_type='full', max_iter=1000, reg_covar=0.1, random_state=42)
gmm_4.fit(df_scaled_full)
labels_gmm_4 = gmm_4.predict(df_scaled_full)

# K-means with subset of features and 5 clusters
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
kmeans.fit(df_scaled_subset)

# add columns with cluster labels
data['cluster_labels_gmm_5'] = labels_gmm_5
data['cluster_labels_gmm_4'] = labels_gmm_4 
data['cluster_labels_kmeans'] = kmeans.labels_
```
